```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Check for missing values and handle them
df['Cuisines'].fillna('Unknown', inplace=True)
df['City'].fillna('Unknown', inplace=True)
```

```python
# Split multiple cuisines in the "Cuisines" column
df['Cuisines'] = df['Cuisines'].str.split(',')

# Explode the cuisines into separate rows
df_exploded = df.explode('Cuisines')
```

```python
df_exploded
```

```python
# Strip whitespace from cuisines and city names
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.strip()
df_exploded['City'] = df_exploded['City'].str.strip()
```

```python
# Distribution of Country Code
country_counts = df_exploded['Country Code'].value_counts()
plt.figure(figsize=(10,5))
country_counts.plot(kind='bar')
plt.title('Distribution of Restaurants by Country Code')
plt.xlabel('Country Code')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Top Cities with the Highest Number of Restaurants
city_counts = df_exploded['City'].value_counts().head(10)
plt.figure(figsize=(10,6))
city_counts.plot(kind='bar')
plt.title('Top 10 Cities with the Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# 3. Top Cuisines
cuisine_counts = df_exploded['Cuisines'].value_counts().head(10)
plt.figure(figsize=(10,6))
cuisine_counts.plot(kind='bar')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.show()
```
