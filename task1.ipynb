```python
rating_counts = df['Aggregate rating'].value_counts()

# Step 2: Visualize the distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Aggregate rating',data=df, palette='viridis')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Step 3: Calculate the percentage distribution of each class
rating_percentage = (rating_counts / len(df)) * 100
#print(rating_percentage)
```
